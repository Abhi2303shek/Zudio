```python
import pandas as pd
```

```python
zudio = pd.read_csv('Zudio_sales_data.csv', sep=',')
zudio
```

```python
zudio.columns
```

```python
zudio = zudio.rename(columns={'Store':'store', 'Country':'country', 'State':'state','City':'city',
                           'Category':'category', 'Clothing Type':'cloth_type', 'Store Number': 'store_no',
                           'Postal Code':'postal' ,'Store Type' :'store_type', 'Store Open Date':'store_open',
                           'Selling Area Size (sq ft)' :'selling_area', 'Store Manager':'manager', 'Manager ID' : 'manager_id',
                           'Store Address' : 'store_add', 'Contact Information' :'contact', 'Operating Hours' : 'op_hours',
                           'Staff Count':'staff', 'Parking Availability' : 'parking', 'Security Features' : 'security',
                           'Order ID' : 'order_id', 'Order Date' : 'order_date', 'Month' : 'month', 'Customer ID' : 'cust_id',
                           'Customer Name' : 'cust_name', 'Product ID' : 'prod_id', 'Price' : 'price', 'Quantity' : 'quantity',
                           'Sales Profit':'profit'})
```

```python
zudio['profit'].dtype
```

```python
zudio['store_open'] = pd.to_datetime(zudio['store_open'])
zudio['store_open'] = zudio['store_open'].dt.date
```

```python
zudio['order_date'] = pd.to_datetime(zudio['order_date'])
zudio['order_date'] = zudio['order_date'].dt.date
```

```python
zudio[['opening_time', 'closing_time']] = zudio['op_hours'].str.split(' - ', expand=True)
```

```python
zudio['opening_time'] = pd.to_datetime(zudio['opening_time'], format='%I %p')
zudio['closing_time'] = pd.to_datetime(zudio['closing_time'], format='%I %p')
```

```python
zudio['operating_hours'] = (zudio['closing_time'] - zudio['opening_time']).dt.total_seconds() / 3600
```

```python
zudio['opening_time'] = zudio['opening_time'].dt.time
```

```python
zudio['closing_time'] = zudio['closing_time'].dt.time
```

```python
zudio = zudio.drop(columns=['store', 'country', 'op_hours'])
```

```python
zudio.to_csv('zudio.csv')
```
